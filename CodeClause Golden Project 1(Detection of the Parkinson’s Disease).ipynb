### Golden Project 1 : Detection of the Parkinson’s Disease

```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('G:/Projects/Codeclause internship/GP(Detection of Parkinson Disease)/Parkinsson disease.xlsx')
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
X = data.drop(['status','name'], axis=1)
y = data['status']
```

```python
X
```

```python
y
```

```python
scaler = MinMaxScaler((-1 , 1))
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2 , random_state=101)
```

```python
# GAUSSIAN NAIVE BAYES(handles negative value better than NAIVE BAYES)

from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train ,y_train)
pred_GNB = model.predict(X_test)
accuracy = accuracy_score(y_test,pred_GNB)
print("accuracy: {}%".format(round(accuracy*100,2)))
```

```python
confusion_mat = confusion_matrix(y_test,pred_GNB)
print("Confusion Matrix GNB:")
columns=['Predicted Healthy', 'Predicted Parkinsons']
index=['True Healthy', 'True Parkinsons']
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', xticklabels=columns, yticklabels=index)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train ,y_train)
pred_LOGR = model.predict(X_test)
accuracy = accuracy_score(y_test,pred_LOGR)
print("accuracy: {}%".format(round(accuracy*100,2)))
```

```python
confusion_mat = confusion_matrix(y_test,pred_LOGR)
print("Confusion Matrix LOGR:")
columns=['Predicted Healthy', 'Predicted Parkinsons']
index=['True Healthy', 'True Parkinsons']
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', xticklabels=columns, yticklabels=index)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# DECISION TREE CLASSIFIER
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(X_train ,y_train)
pred_DTC = clf.predict(X_test)
accuracy = accuracy_score(y_test,pred_DTC)
print("accuracy: {}%".format(round(accuracy*100,2)))
```

```python
confusion_mat = confusion_matrix(y_test,pred_DTC)
print("Confusion Matrix DTC:")
columns=['Predicted Healthy', 'Predicted Parkinsons']
index=['True Healthy', 'True Parkinsons']
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', xticklabels=columns, yticklabels=index)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(random_state=42)
RFC.fit(X_train ,y_train)
pred_RFC = RFC.predict(X_test)
accuracy = accuracy_score(y_test,pred_RFC)
print("accuracy: {}%".format(round(accuracy*100,2)))
```

```python
confusion_mat = confusion_matrix(y_test,pred_RFC)
print("Confusion Matrix RFC:")
columns=['Predicted Healthy', 'Predicted Parkinsons']
index=['True Healthy', 'True Parkinsons']
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', xticklabels=columns, yticklabels=index)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# SUPPORT VECTOR MACHINE
from sklearn.svm import SVC

svmc = SVC(kernel='linear')
svmc.fit(X_train ,y_train)
pred_svmc = svmc.predict(X_test)
accuracy = accuracy_score(y_test,pred_svmc)
print("accuracy: {}%".format(round(accuracy*100,2)))
```

```python
confusion_mat = confusion_matrix(y_test,pred_svmc)
print("Confusion Matrix svmc:")
columns=['Predicted Healthy', 'Predicted Parkinsons']
index=['True Healthy', 'True Parkinsons']
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', xticklabels=columns, yticklabels=index)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# K-NEAREST NEIGHBORS

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 8,p=2,metric ='minkowski')
knn.fit(X_train,y_train)
pred_knn = knn.predict(X_test)
accuracy = accuracy_score(y_test,pred_knn)
print("accuracy: {}%".format(round(accuracy*100,2)))
```

```python
confusion_mat = confusion_matrix(y_test,pred_knn)
print("Confusion Matrix knn:")
columns=['Predicted Healthy', 'Predicted Parkinsons']
index=['True Healthy', 'True Parkinsons']
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', xticklabels=columns, yticklabels=index)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
